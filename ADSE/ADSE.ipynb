```python
govEx = pd.read_csv('government expentiture on edu.csv')
govEx
```

```python
plt.figure(figsize=(12,10))
kolom = len(govEx.columns)
negara = govEx['Country Name'].unique()
for i in range (10):
    
    row = govEx.iloc[i]
    arr =[]
    for j in range(16,kolom):
        arr.append(row.iloc[j])
    
    sns.lineplot(x=[2010,2011,2012,2013,2014,2015,2016,2017],y=arr)
plt.legend(labels=negara)    
```
